## Пример воспроизводимого решения

Если вы претендуете на призовое место в финальном этапе хакатона, то от вас требуется не только загрузить в систему Яндекс.Контест качественные прогнозы модели, но и продемонстрировать, как эти прогнозы были получены.

Для этого от вас требуется оформить Jupyter-ноутбук, демонстрирующий получение ваших прогнозов. Ноутбук и сопутствующие файлы должны быть размещены в папке на Яндекс.Диске (не забудьте включить доступ к папке по ссылке!), а ноутбук должен соответствовать некоторым критериям:
* в ноутбуке должен быть продемонстрирован перечень библиотек для языка Python, которые установлены на вашем компьютере, и их версии
* в решении должен быть зафиксирован Random Seed (подробнее об этом - ниже), чтобы нивелировать элемент случайности и дать возможность организаторам получить точь-в-точь такое же решение, как у вас
* при **последовательном** выполнении ячеек ноутбука должен сформироваться файл с точно таким же прогнозом, который был загружен Вами в Контест

Текущий ноутбук является примером, удовлетворяющим перечисленным критерям.

Давайте пройдём по нему по порядку:

<br><br>

**Первый обязательный критерий:** мы выведем на экран список всех библиотек из нашего окружения Python с помощью команды `!pip list`. Это требуется для того, чтобы у организаторов была возможность воспроизвести решение на том же наборе библиотек, что и у Вас. 

```python
!pip list
```

<br><br>

```python
import pandas as pd
import numpy as np
```

<br><br>

**Второй обязательный критерий:** фиксация random seed. Random seed – это числовое значение, используемое для инициализации генератора псевдослучайных чисел. Некоторые алгоритмы машинного обучения несут в себе элемент случайности: от запуска к запуску итоговый результат может несколько отличаться. Фиксируя перед вычислениями random seed мы гарантируем, что при повторном запуске решения мы получим точно такой же результат, какой получили ранее.

```python
np.random.seed(2026)
```

<br><br>

**Третий обязательный критерий:** наличие кода, воспроизводящего итоговую посылку в Контест.

В нашем случае давайте построим очень простую модель: воспроизведём бейзлайн первого этапа.

```python
# загрузим имеющиеся у нас данные
# Внимание: чтобы эта ячейка отработала, файл train.csv должен находиться в той же директории, что и этот ноутбук
train = pd.read_csv('train_2.csv')
```

```python
# сформируем список магазинов, для которых надо построить прогноз
shops_list = train['new_id'].unique()
print('Число магазинов:', len(shops_list))
```

Из ноутбука с примером формирования решения в корректном формате первого этапа хакатона возьмём среднее значение РТО магазинов за октябрь.

```python
avg_rto_oct = 33503980.259272862
```

**Несколько подсказок:** Обратим внимание, что число магазинов в данных второго этапа отличается от числа магазинов первого этапа. Это говорит о том, что выборки разные. Скорее всего, подсчитав среднее значение РТО за октябрь для магазинов текущей выборки, мы получим другой прогноз, который с высокой степенью вероятности превзойдёт текущий по качеству.

Кроме того, теперь у нас есть данные за последующие месяцы. Мы хотим прогнозировать РТО за март 2025, а значит, среднее за февраль 2025 года может показать более точный прогноз.

Этот ноутбук - лишь демонстрация формата сдачи решения, где используется простейшая бейзлайн-модель. Загрузка прогнозов бейзлайн-модели не принесёт баллов. Но используя подсказки выше, Вы можете мгновенно построить другие простейшие модели, которые тем не менее будут лучше, чем бейзлайн.

```python
# Создадим объект DataFrame (аналог таблицы)
sample_submission = pd.DataFrame()
# Добавим столбец 'new_id' с идентификаторами магазинов
sample_submission['new_id'] = shops_list
# Добавим столбец 'rto', куда поместим значение нашего прогноза
sample_submission['rto'] = avg_rto_oct
```

```python
sample_submission.to_csv('test.csv', index=None)
```

Ноутбук, соответствующий перечисленным выше критерям, требуется разместить в каком-нибудь облачном хранилище (например, на Яндекс.Диске), и во второй задаче финального этапа оставить ссылку на это хранилище (предварительно убедитесь, что в настройках открыт доступ к директории по ссылке).
